```python
import pandas as pd
import os
import ast
```

```python
def getAllFilenames(dirPath, exclude=[]):
    filenames = []
    for filename in os.listdir(dirPath):
        if filename not in exclude:
            filenames.append(filename)
    return filenames
```

```python
filenames = getAllFilenames("/share/GithubAnalysis/pullFiles/", ["combined", "Pulls_VS_Total.csv"])
```

```python
def getValue(payload, path):
    root = payload
    for pathItem in path:
        if pathItem not in root:
            return None
        root = root[pathItem]
    return root
        
def pickRowData(df, doc):
    payload = ast.literal_eval(df["payload"])
    doc["event_time"].append(df["created_at"])
    doc["user_login"].append(getValue(payload, ["pull_request", "user", "login"]))
    doc["forks_count"].append(getValue(payload, ["pull_request", "base", "repo", "forks_count"]))
    doc["language"].append(getValue(payload, ["pull_request", "base", "repo", "language"]))
    doc["has_wiki"].append(getValue(payload, ["pull_request", "base", "repo", "has_wiki"]))
    doc["watchers_count"].append(getValue(payload, ["pull_request", "base", "repo", "watchers_count"]))
    doc["size"].append(getValue(payload, ["pull_request", "base", "repo", "size"]))
    doc["additions"].append(getValue(payload, ["pull_request", "additions"]))
    doc["deletions"].append(getValue(payload, ["pull_request", "deletions"]))
    doc["changed_files"].append(getValue(payload, ["pull_request", "changed_files"]))
    doc["number_commits"].append(getValue(payload, ["pull_request", "commits"]))
    doc["created_at"].append(getValue(payload, ["pull_request", "base", "repo", "created_at"]))
    doc["open_issues_count"].append(getValue(payload, ["pull_request", "base", "repo", "open_issues_count"]))
    doc["name"].append(getValue(payload, ["pull_request", "base", "repo", "name"]))
    doc["description"].append(getValue(payload, ["pull_request", "base", "repo", "description"]))
    doc["body"].append(getValue(payload, ["pull_request", "body"]))
    if df["type"] == "PullRequestReviewCommentEvent":
        doc["secondary_user"].append(getValue(payload, ["comment", "user", "login"]))
    else:
        doc["secondary_user"].append(None)
    
doc = {
    "event_time": [],
    "user_login": [],
    "secondary_user": [],
    "forks_count": [],
    "language": [],
    "has_wiki": [],
    "watchers_count": [],
    "size": [],
    "additions": [],
    "deletions": [],
    "changed_files": [],
    "number_commits": [],
    "created_at": [],
    "open_issues_count": [],
    "name": [],
    "description": [],
    "body": []
}
for filename in filenames:
    print(filename)
    data = pd.read_csv("/share/GithubAnalysis/pullFiles/" + filename)
    for index, row in data.iterrows():
        pickRowData(row, doc)
allData = pd.DataFrame(doc)
allData
```
