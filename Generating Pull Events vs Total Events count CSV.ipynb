```python
import pandas as pd
```

```python
pull_info = []

def count_pulls(day):
    global pull_info
    for hour in range (0,24):
        dset = pd.read_json("/share/GithubAnalysis/2016-12-"+ day +"-"+str(hour)+".json", lines = True)
        numTotal = len(dset)
        dpulls = dset[(dset['type'] == 'PullRequestReviewCommentEvent') | (dset['type'] == 'PullRequestEvent')]
        numPulls = len(dpulls)
        pull_info.append((int(day), hour, numPulls, numTotal))
        
for day in range (1,32):
    if day < 10:
        day = '0'+str(day)
    else:
        day = str(day)
    count_pulls(day)
```

```python
df = pd.DataFrame(pull_info, columns = ['day', 'hour', 'pull_events', 'total_events'])
```

```python
df['percent_pulls'] = df['pull_events']/df['total_events']
```
