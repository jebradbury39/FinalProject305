We want to have 0.5 million rows, where each row corresponds to an event. Each row requires 2 api calls. 1000000 / 31 = 32259 calls per day, so 32259 / 24 = 1345 calls per hour, so a random sample of 672 events from every hour.

```python
import calendar
import time
import os
import json
```

```python
clientApps = []
currentAppIndex = 0
#register each app
for i in range(len(auth)):
    clientApp = "client_id=" + auth["Client ID"].loc[i] + "&client_secret=" + auth["Client Secret"].loc[i]
    clientApps.append({
            "target": clientApp,
            "resetTime": None,
            "callsRemaining": 5000
        })
```

```python
def findNextApp():
    global clientApps
    global currentAppIndex
    
    count = 0
    found = (False, clientApps[currentAppIndex]["resetTime"]) #Did we find one, and if not, what is min reset time?
    currentTime = calendar.timegm()
    while True:
        count += 1
        if (count == len(clientApps)):
            break
            
        if currentAppIndex == len(clientApps) - 1:
            currentAppIndex = 0
        else:
            currentAppIndex += 1
            
        clientApp = clientApps[currentAppIndex]
        if clientApp["resetTime"] is None:
            found[0] = True
            break
            
        if clientApp["resetTime"] is not None:
            if found[1] > clientApp["resetTime"]:
                found[1] = clientApp["resetTime"]
            if clientApp["resetTime"] > currentTime:
                found[0] = True
                break
    found[1] -= currentTime #To give the the number of seconds to sleep
    
def updateRemaining(clientApp):
    reqRate = requests.get("https://api.github.com/rate_limit?" + clientApp["target"])
    clientApp["callsRemaining"] = reqRate.json()["resources"]["core"]["remaining"]
    
def makeCall(url):
    global clientApps
    global currentAppIndex
    time.sleep(0.1)
    clientApp = clientApps[currentAppIndex]
    updateRemaining(clientApp)
    if (clientApp["callsRemaining"] == 0):
        req = requests.get("https://api.github.com/rate_limit?" + clientApp["target"])
        clientApp["resetTime"] = req.json()["resources"]["core"]["reset"]
        found = findNextApp()
        if found[0] == False:
            time.sleep(found[1])
    clientApp = clientApps[currentAppIndex]
    req = requests.get(url + "?" + clientApp["target"])
    clientApp["callsRemaining"] -= 1
    
    res = req.json()
    if "location" not in res:
        res["location"] = None
    if "created_at" not in res:
        res["created_at"] = None
    if "login" not in res:
        res["login"] = None
    if "name" not in res:
        res["name"] = None
    if "company" not in res:
        res["company"] = None
    if "hireable" not in res:
        res["hireable"] = None
    if "language" not in res:
        res["language"] = None
    if "size" not in res:
        res["size"] = None
    return res
```

```python
# we need to loop every hour, obtaining 672 randoms from each
docs = {}
path = "/share/GithubAnalysis"
filenames = []
for filename in os.listdir(path):
    if filename != "called":
        filenames.append(filename)
            
            
data = pd.Series(filenames)
data
```

```python
for i in range(len(filenames)):
    filename = filenames[i]
    data = pd.read_json("/share/GithubAnalysis/" + filename, lines=True)
    sample = data.sample(n=672)
    
    print("\n" + str(i) + ": " + filename)
    langSeries = []
    sizeSeries = []
    
    locationSeries = []
    createdAtSeries = []
    loginSeries = []
    nameSeries = []
    companySeries = []
    hireableSeries = []
    
    for j in range(len(sample)):
        sys.stdout.write("\r - %i" % j)
        sys.stdout.flush()
        
        actor = sample["actor"].iloc[j]
        actorReq = makeCall(actor["url"])
        locationSeries.append(actorReq["location"])
        createdAtSeries.append(actorReq["created_at"])
        loginSeries.append(actorReq["login"])
        nameSeries.append(actorReq["name"])
        companySeries.append(actorReq["company"])
        hireableSeries.append(actorReq["hireable"])
        
        repo = sample["repo"].iloc[j]
        repoReq = makeCall(repo["url"])
        langSeries.append(repoReq["language"])
        sizeSeries.append(repoReq["size"])
        
    seriesDf = pd.DataFrame({
        "language": langSeries,
        "size": sizeSeries,
        "location": locationSeries,
        "created_at": createdAtSeries,
        "login": loginSeries,
        "name": nameSeries,
        "company": companySeries,
        "hireable": hireableSeries
    })
    seriesDf.to_csv("/share/GithubAnalysis/called/" + filename.rstrip(".json") + ".csv", sep=",")
    break
    
```
